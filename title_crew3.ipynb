```python
from datetime import datetime
from datetime import timedelta
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://datasets.imdbws.com/title.crew.tsv.gz"
df1 = pd.read_csv(url, sep = '\t')
```

```python
print(df1.info())
```

```python
display(df1.head())
```

```python
print(df1.isnull().sum())
```

```python
# Remplacer les '\N' par NaN
df1.replace("\\N", pd.NA, inplace=True)

# Supprimer les lignes contenant des NaN
df_cleaned = df1.dropna()

# Afficher le résultat
display(df_cleaned)
```

```python
display(df1.head())
```

```python
url = "https://datasets.imdbws.com/title.episode.tsv.gz"
df2 = pd.read_csv(url, sep = '\t')
```

```python
display(df2.head())
```

```python
print(df2.isnull().sum())
```

```python
# Fusionner les deux DataFrames sur la colonne 'tconst'
df_crew = df1.merge(df2, on='tconst', how='inner')
display(df_crew.head())
```

```python
# Remplacer les '\N' par NaN
df_crew.replace("\\N", pd.NA, inplace=True)

# Supprimer les lignes contenant des NaN
df_cleaned = df_crew.dropna()

# Afficher le résultat
display(df_cleaned)
```

```python
df_cleaned.shape
```

```python
# Supprimer les doublons dans df_cleaned
df_cleaned = df_cleaned.drop_duplicates()

# Afficher le résultat
display(df_cleaned)
```

```python
# Renommer le DataFrame
df_crew_cleaned = df_cleaned

# Supprimer l'ancien DataFrame si vous n'en avez plus besoin
del df_cleaned  # Facultatif

# Vérifier le nouveau DataFrame
display(df_crew_cleaned)
```
