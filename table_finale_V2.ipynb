```python
# Convertir chaque élément de la colonne 'genres' en une chaîne de caractères si nécessaire
# Supposons que chaque élément est une chaîne de caractères ou une liste de chaînes de caractères
# Vous pouvez ajuster la logique de conversion en fonction du contenu réel

def convert_genres(genres):
    if isinstance(genres, list):
        return ' '.join(genres)
    elif isinstance(genres, str):
        return genres
    else:
        return str(genres)

df_tmdb['genres'] = df_tmdb['genres'].apply(convert_genres)
# Vérifier que la conversion a réussi
print(df_tmdb['genres'].head())
```

```python
import requests
import pandas as pd
from IPython.display import display, Image

# Configuration API
API_KEY = "ed5d65da6eaa9f2dec411dae1e19302d"
BASE_URL = "https://api.themoviedb.org/3"

# Fonction pour obtenir les détails d'un film
def get_movie_details(movie_id, api_key):
    url = f"{BASE_URL}/movie/{movie_id}?api_key={api_key}"
    response = requests.get(url)
    return response.json()

# Fonction pour construire l'URL de l'image
def get_image_url(poster_path, file_size='w200'):
    base_url = "https://image.tmdb.org/t/p/"
    return f"{base_url}{file_size}{poster_path}"

# Fonction pour récupérer l'affiche d'un film et construire l'URL de l'image
def get_movie_poster_url(movie_id, api_key, file_size='w154'):
    movie_details = get_movie_details(movie_id, api_key)
    poster_path = movie_details.get('poster_path', '')
    return get_image_url(poster_path, file_size) if poster_path else None

# Liste d'IDs de films
movie_ids = ['tt0035423', 'tt0097106', 'tt0097318']

# Récupérer les URLs des affiches pour les 10 premiers films
poster_urls = [get_movie_poster_url(movie_id, API_KEY) for movie_id in movie_ids]

# Afficher les affiches
for url in poster_urls:
    if url:
        display(Image(url=url))

```
