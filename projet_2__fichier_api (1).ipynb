```python
import requests
import pandas as pd
import time

# Configuration de l'API
API_KEY = "ed5d65da6eaa9f2dec411dae1e19302d"
BASE_URL = "https://api.themoviedb.org/3"
PAGE_LIMIT = 500  # Limite de pages à récupérer (jusqu'à 500 pour obtenir beaucoup de films)
MOVIES = []

# Fonction pour récupérer les films sans filtre de langue
def fetch_movies(page):
    url = f"{BASE_URL}/discover/movie?api_key={API_KEY}&page={page}"  # Pas de filtre sur la langue ici
    try:
        response = requests.get(url)
        response.raise_for_status()  # Si la réponse a un code d'erreur, une exception sera levée
        return response.json()['results']
    except requests.exceptions.RequestException as e:
        print(f"Erreur lors de la récupération des films (page {page}): {e}")
        return []

# Récupérer les films de plusieurs pages
for page in range(1, PAGE_LIMIT + 1):
    print(f"Récupération de la page {page}...")
    movies = fetch_movies(page)
    if not movies:
        break  # Si la page ne contient pas de films, on arrête
    MOVIES.extend(movies)
    time.sleep(1)  # Pause pour éviter d'envoyer trop de requêtes rapidement

# Vérifie combien de films ont été récupérés
print(f"{len(MOVIES)} films récupérés.")

# Sauvegarder les films récupérés dans un fichier CSV
if MOVIES:
    file_path = '/content/filtered_movies_10000.csv'
    df = pd.DataFrame(MOVIES)
    df.to_csv(file_path, index=False)
    print(f"{len(df)} films sauvegardés dans '{file_path}'")
else:
    print("Aucun film récupéré.")
```
