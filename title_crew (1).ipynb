```python
from datetime import datetime
from datetime import timedelta
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://datasets.imdbws.com/title.crew.tsv.gz"
df1 = pd.read_csv(url, sep = '\t')
```

```python
print(df1.info())
```

```python
display(df1.head())
```

```python
print(df1.isnull().sum())
```

```python
df1.replace("\\N", "Nan", inplace=True)
```

```python
display(df1.head())
```

```python
url = "https://datasets.imdbws.com/title.episode.tsv.gz"
df2 = pd.read_csv(url, sep = '\t')
```

```python
display(df2.head())
```

```python
print(df2.isnull().sum())
```

```python
df2.replace("\\N", "Nan", inplace=True)
```

```python
display(df2.head())
```

```python
# Fusionner les deux DataFrames sur la colonne 'tconst'
df_crew = df1.merge(df2, on='tconst', how='inner')
```

```python
# Afficher le DataFrame fusionn
display(df_crew.head(5))
```
