```python
#renommer la colonne "titleId" par "tconst"
df3 = df3.rename(columns={'titleId': 'tconst'})
df3.head(2)
```

```python
# Fusionner les DataFrames 
df_merged = pd.merge(df1, df2, on='tconst', how='outer')
df_merged = pd.merge(df_merged, df3, on='tconst', how='outer')
df_merged = pd.merge(df_merged,df4, on='tconst', how='outer')
df_merged = pd.merge(df_merged,df5, on='nconst', how='outer')
```
