```python
from datetime import datetime
from datetime import timedelta
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = "https://datasets.imdbws.com/title.ratings.tsv.gz"
df7 = pd.read_csv(url, sep = '\t')
```

```python
# Vérification des doublons basés sur les colonnes 'tconst'
duplicates = df7.duplicated(subset=['tconst'])
print("Nombre de doublons :", duplicates.sum())
```

```python
df7.head(15)
```

```python
# Trier le DataFrame par 'averageRating' en ordre décroissant
df7_sorted = df7.sort_values(by='averageRating', ascending=False)

# Afficher le DataFrame trié
print("Données triées par averageRating (décroissant) :")
print(df7_sorted)
```

```python
# Compter le nombre de 'tconst' pour chaque valeur unique de 'averageRating'
rating_counts = df7.groupby('averageRating').size().reset_index(name='count')

# Afficher le DataFrame avec les comptages
print("Nombre de tconst pour chaque classe averageRating :")
display(rating_counts.head(50))
```
