```python
from datetime import datetime
from datetime import timedelta
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import gdown
import pandas as pd
```

```python
# ID du fichier Google Drive pour le deuxième fichier
file_id_title_ratings = '1qJytCbyZVgT7HoI8a4t90QG06JNKa13I'
title_ratings = f'https://drive.google.com/uc?id={file_id_title_ratings}'
output_title_ratings = 'title_ratings.csv'

# Télécharger le fichier CSV depuis Google Drive
gdown.download(title_ratings, output_title_ratings, quiet=False)

# Lire le fichier CSV avec pandas
df1 = pd.read_csv(output_title_ratings, delimiter=',', on_bad_lines='skip', engine='python')
```

```python

# l' ID du fichier Google Drive
file_id_title_principals = '1Lx4VzThPS8SZlSb7JWrz7Rb0O1Jg-Z8-'
title_principals = f'https://drive.google.com/uc?id={file_id_title_principals}'
output_title_principals = 'title_principals_final.csv'

# Télécharger le fichier CSV depuis Google Drive
gdown.download(title_principals, output_title_principals, quiet=False)

# Lire le fichier CSV avec pandas
df2 = pd.read_csv(output_title_principals, delimiter=',', on_bad_lines='skip', engine='python')
df2.head(2)
```

```python
file_id_titleakas = '1miW6rfrC8avh1nnm7WjD2cRO-_atIQNj'
titleakas = f'https://drive.google.com/uc?id={file_id_titleakas}'
output_titleakas = 'titleakas.csv'

# Télécharger le fichier CSV depuis Google Drive
gdown.download(titleakas, output_titleakas, quiet=False)

# Lire le fichier CSV avec pandas
df3 = pd.read_csv(output_titleakas, delimiter=',', on_bad_lines='skip', engine='python')
```

```python
#renommer la colonne "titleId" par "tconst"
df3 = df3.rename(columns={'titleId': 'tconst'})
df3.head()
```

```python
# ID du fichier Google Drive pour le nouveau fichier
file_id_title_basics = '1LQMg069N6kP7AzxgoglkyPaNizaMDMBt'
title_basics = f'https://drive.google.com/uc?id={file_id_title_basics}'
output_title_basics = 'title_basics.csv'

# Télécharger le fichier CSV depuis Google Drive
gdown.download(title_basics, output_title_basics, quiet=False)

# Lire le fichier CSV avec pandas
df4 = pd.read_csv(output_title_basics, delimiter=',', on_bad_lines='skip', engine='python')
df4.head()
```

```python
import pandas as pd

# URL du fichier Google Drive pour le nouveau fichier
file_id_name_basics = '1UVpk5usik_flBdg1FjFeU8WM7bb81vw5'
name_basics = f'https://drive.google.com/uc?id={file_id_name_basics}'

# Lire le fichier CSV avec pandas
try:
    df5 = pd.read_csv(name_basics, delimiter=',', on_bad_lines='skip', engine='python')
    if df5.empty:
        print("Le fichier est vide.")
    else:
        display(df5.head())
except pd.errors.EmptyDataError:
    print("Aucune donnée à analyser dans le fichier.")
except Exception as e:
    print(f"Une erreur est survenue : {e}")

```

```python
# Vérifier les colonnes de chaque DataFrame avant la fusion
display(df1.columns)
display(df2.columns)
display(df3.columns)
display(df4.columns)
display(df5.columns)
```

```python
#fusion de df1 et df2 sur la colonne'tcont'
df_merged_1 = pd.merge(df1, df2, on=['tconst'], how='outer')
df_merged_1.head(2)
```

```python
#fusion de df_merged_1 et df3 sur la colonne'tcont'
#supression de la colonne 'ordering' dans df3 avant la fusion
df3_dropped = df3.drop(columns=['ordering'])
df_merged_2 = pd.merge(df_merged_1, df3_dropped, on=['tconst'], how='outer')
df_merged_2.head(30)
```

```python
#fusion de df_merged_2 et df4 sur la colonne'tcont'
df_merged_3 = pd.merge(df_merged_2, df4, on=['tconst'], how='outer')
df_merged_3.head(2)
```

```python
#fusion de df_merged_3 et df2 sur la colonne'ncont'
df_merged_4 = pd.merge(df_merged_3, df5, on=['nconst'], how='outer')
df_merged_4.head(15)
```

```python
# Calculer le nombre de doublons basés sur la colonne 'tconst' et la colonne 'nconst'
num_duplicates = df_merged_4.duplicated(subset=['tconst', 'nconst']).sum()
print(f"Nombre de doublons : {num_duplicates}")
```

```python
# Supprimer les doublons basés sur la colonne 'tconst' et la colonne 'nconst'
merged_df = df_merged_4.drop_duplicates(subset=['tconst', 'nconst'])
display(merged_df.head(35))
```

```python
#vérifier s'il y a des doublons : 
# Calculer le nombre de doublons basés sur la colonne 'tconst' et la colonne 'nconst'
check_duplicates = merged_df.duplicated(subset=['tconst', 'nconst']).sum()
print(f"Nombre de doublons : {check_duplicates}")
```

```python
# Obtenir les valeurs uniques dans la colonne 'category'
unique_category = merged_df['category'].unique()
print(unique_category)
```

```python
# Calculer le nombre de valeurs NaN dans la colonne 'title'
num_nan_title = merged_df['title'].isna().sum()

print(f"Nombre de valeurs NaN dans la colonne 'title' : {num_nan_title}")
```

```python
# Supprimer les lignes contenant des valeurs NaN dans la colonne 'title'
df_cleaned = merged_df.dropna(subset=['title'])

display(df_cleaned.columns)
```
