```python
import gdown
import pandas as pd

# ID du fichier Google Drive
file_id = '1WzrI9WTXCPRhIdaxs7jbsPjZAMxlMRW1'

# URL de téléchargement direct
url = f'https://drive.google.com/uc?id={file_id}'

# Télécharger le fichier et le sauvegarder en tant que 'file.csv'
output = 'file.csv'
gdown.download(url, output, quiet=False)

# Lire le fichier CSV avec pandas
df = pd.read_csv(output)

# Afficher les premières lignes du DataFrame
display(df.head())
```

```python
display(df['genres'])
```

```python
# Dictionnaire de synonymes
synonyms = {
    'Sci-Fi': 'Science Fiction',
    'Horror': 'Horror',
    'Action': 'Action',
    'Comedy': 'Comedy',
    'Drama': 'Drama',
    'Romance': 'Romance',
    'Thriller': 'Thriller',
    'Animation': 'Animation',
    'Documentary': 'Documentary',
    'History': 'History'
    # Ajoute plus de synonymes si nécessaire
}

# Fonction pour nettoyer et dédupliquer les genres
def clean_genres(genres):
    # Diviser par la virgule
    split_genres = genres.split(',')
    # Remplacer les synonymes et nettoyer les genres
    cleaned_genres = [synonyms.get(g.strip(" []'"), g.strip(" []'")) for g in split_genres if g.strip(" []'")]
    # Utiliser un ensemble pour supprimer les doublons
    unique_genres = set(cleaned_genres)
    # Rejoindre les genres propres par des virgules
    return ', '.join(sorted(unique_genres))

# Appliquer la fonction à la colonne 'genres'
df['genres_clean'] = df['genres'].apply(clean_genres)

# Afficher le DataFrame nettoyé
display(df[['genres', 'genres_clean']])
```

```python
# Supprimer la colonne 'genres'
df.drop(columns=['genres'], inplace=True)

# Renommer la colonne 'genres_clean' en 'genres'
df.rename(columns={'genres_clean': 'genres'}, inplace=True)

# Afficher le DataFrame final
display(df.head(2))
```

```python
# Fonction pour nettoyer et dédupliquer les languages
def clean_languages(languages):
    # Diviser par la virgule
    split_languages = languages.split(',')
    # Remplacer les synonymes et nettoyer les languages
    cleaned_languages = [synonyms.get(lang.strip(" []'"), lang.strip(" []'")) for lang in split_languages if lang.strip(" []'")]
    # Utiliser un ensemble pour supprimer les doublons
    unique_languages = set(cleaned_languages)
    # Rejoindre les languages propres par des virgules
    return ', '.join(sorted(unique_languages))

# Appliquer la fonction à la colonne 'languages'
df['languages_clean'] = df['languages'].apply(clean_languages)

# Supprimer la colonne 'languages'
df.drop(columns=['languages'], inplace=True)

# Renommer la colonne 'languages_clean' en 'languages'
df.rename(columns={'languages_clean': 'languages'}, inplace=True)

# Afficher le DataFrame final
display(df['languages'].head())
```
